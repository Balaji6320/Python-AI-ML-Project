```python
selected_columns =['Age','Gender','Daily_Screen_Time(hrs)','Sleep_Quality(1-10)',
                   'Stress_Level(1-10)','Days_Without_Social_Media',
                   'Exercise_Frequency(week)','Social_Media_Platform','Happiness_Index(1-10)']
selected_data = data[selected_columns].copy()
```

```python
# Features and target
x=selected_data.drop("Happiness_Index(1-10)", axis=1)     # independent variables
y=selected_data["Happiness_Index(1-10)"]                  # dependent variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
numerical_features = ['Age','Daily_Screen_Time(hrs)','Sleep_Quality(1-10)',
                    'Stress_Level(1-10)','Days_Without_Social_Media','Exercise_Frequency(week)']
nominal_features = ['Gender','Social_Media_Platform']
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('ohe', OneHotEncoder(), nominal_features)
    ])
```

### Linear Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
Linear_Regression = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", LinearRegression())
])
```

```python
Linear_Regression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_transformed = preprocessor.fit_transform(x_train)
```

```python
X_transformed[0]
```

```python
Linear_Regression.fit(x, y)
```

```python
x_test['Happiness_Index(1-10)']=y_test
x_test['Predicted_Happiness_Index']=Linear_Regression.predict(x_test)
```

```python
x_test
```

```python
x_test['actual_error']=x_test['Happiness_Index(1-10)'] - x_test['Predicted_Happiness_Index']
x_test['absolute_error']=abs(x_test['actual_error'])
x_test['absolute_percentage_error']=abs(x_test['actual_error']/x_test['Predicted_Happiness_Index'])
```

```python
x_test['absolute_percentage_error'].mean()
```

### Random forest

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.ensemble import RandomForestRegressor
random_forest = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model',  RandomForestRegressor(
        n_estimators=300,
        max_depth=15,
        min_samples_split=5,
        random_state=42))
])
random_forest.fit(x_train, y_train)
```

```python
x_test['Happiness_Index(1-10)']=y_test
x_test['Predicted_Happiness_Index']= random_forest.predict(x_test)
x_test['actual_error']=x_test['Happiness_Index(1-10)'] - x_test['Predicted_Happiness_Index']
x_test['absolute_error']=abs(x_test['actual_error'])
x_test['absolute_percentage_error']=abs(x_test['actual_error']/x_test['Predicted_Happiness_Index'])
```

```python
x_test['absolute_error'].mean()
```

### XGBoost

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from xgboost import XGBRegressor
XGB = Pipeline(steps=[
('preprocessor', preprocessor),
('model', XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
))
])
XGB.fit(x_train, y_train)
```

```python
x_test['Happiness_Index(1-10)']=y_test
x_test['Predicted_Happiness_Index']= XGB.predict(x_test)
x_test['actual_error']=x_test['Happiness_Index(1-10)'] - x_test['Predicted_Happiness_Index']
x_test['absolute_error']=abs(x_test['actual_error'])
x_test['absolute_percentage_error']=abs(x_test['actual_error']/x_test['Predicted_Happiness_Index'])
```

### Light Gradient Boosting Machine

```python
from lightgbm import LGBMRegressor

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
LGB = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LGBMRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=-1,
        random_state=42))
])
LGB.fit(x_train, y_train)
```

```python
x_test['Happiness_Index(1-10)']=y_test
x_test['Predicted_Happiness_Index']= LGB.predict(x_test)
x_test['actual_error']=x_test['Happiness_Index(1-10)'] - x_test['Predicted_Happiness_Index']
x_test['absolute_error']=abs(x_test['actual_error'])
x_test['absolute_percentage_error']=abs(x_test['actual_error']/x_test['Predicted_Happiness_Index'])
```
